points属性

```python
points = []
points_list= []
for i in pr['points']:
    points.append(i)
    if(i==i):
        points_list.append(i)
values=0
print("均值：",np.mean(points_list))
print("中位数：",np.median(points_list))
print("最小值：",min(points_list))
print("最大值：",max(points_list))
for i in points:
    if(i!=i):
        values=values+1
print("缺失值个数:",values)
print("四分位数:",np.percentile(points_list, (25, 50, 75), interpolation='midpoint'))
counts = np.bincount(points_list)
# 返回众数
print("众数：",np.argmax(counts))
```

直方图

```python
def draw_hist_points(myList,Title,Xlabel,Ylabel,Xmin,Xmax,Ymin,Ymax):
      pl.hist(myList,100)
      pl.xlabel(Xlabel)
      pl.xlim(Xmin,Xmax)
      pl.ylabel(Ylabel)
      pl.ylim(Ymin,Ymax)
      pl.title(Title)
      pl.show()
data = points_list
draw_hist_points(points, 'points', 'points', 'number', min(data), 100, 0.0, 30000)
```

数值的最大值是100，把横坐标分成100块,画出直方图。

# 用最高频率值来填补缺失值

points属性

```python
temp1 = pr.copy(deep = True);
for i in temp1.columns:
    DataColumn = temp1[i];  # 获取该列
    MostFrequentElement = DataColumn.value_counts().idxmax();
    # print('属性', i, '的众数是', str(MostFrequentElement));
    DataColumn = DataColumn.fillna(value=MostFrequentElement);  # 众数填补缺失值
    temp1[i] = DataColumn;
```

```python
points2=[]
for i in temp1['points']:
    points2.append(i)
data = points2
draw_hist_points(points2, 'price', 'price', 'number', min(data), 100, 0.0, 30000)
```

```python
stats.probplot(points2, dist="norm", plot=pl)
pl.show()
```

```python
data = pd.DataFrame({"points": points2})
data.boxplot()  
pl.ylabel("points")
pl.xlabel("points")
pl.show()
```

price属性

```python
temp2 = pr.copy(deep = True);
for i in temp2.columns:
    DataColumn = temp2[i];  # 获取该列
    MostFrequentElement = DataColumn.value_counts().idxmax();
    # print('属性', i, '的众数是', str(MostFrequentElement));
    DataColumn = DataColumn.fillna(value=MostFrequentElement);  # 众数填补缺失值
    temp2[i] = DataColumn;
```

```python
price2=[]
for i in temp1['price']:
    price2.append(i)
data = price2
draw_hist_price(price2, 'price', 'price', 'number', min(data), 500, 0.0, 10000)
```
