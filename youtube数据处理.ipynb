# 数值属性

views属性

```python
views = []
views1= []
for i in pr['views']:
    views.append(i)
    if(i==i):
        views1.append(i)
values=0
print("均值：",np.mean(views1))
print("中位数：",np.median(views1))
print("最小值：",min(views1))
print("最大值：",max(views1))
for i in views:
    if(i!=i):
        values=values+1
print("缺失值个数:",values)
print("四分位数:",np.percentile(views1, (25, 50, 75), interpolation='midpoint'))
counts = np.bincount(views1)
print("众数：",np.argmax(counts))
```

盒图

```python
data = pd.DataFrame({"views": views})
data.boxplot()  
pl.ylabel("views")
pl.show()
```

直方图

```python
def draw_hist(myList,Title,Xlabel,Ylabel,Xmin,Xmax,Ymin,Ymax):
      pl.hist(myList,2300)
      pl.xlabel(Xlabel)
      pl.xlim(Xmin,Xmax)
      pl.ylabel(Ylabel)
      pl.ylim(Ymin,Ymax)
      pl.title(Title)
      pl.show()
data = views
draw_hist(views, 'comment_count', 'comment_count', 'number', min(data), 3000000, 0.0,5000)
```
