viewsÂ±ûÊÄß

```python
views = []
views1= []
for i in pr['views']:
    views.append(i)
    if(i==i):
        views1.append(i)
values=0
print("ÂùáÂÄºÔºö",np.mean(views1))
print("‰∏≠‰ΩçÊï∞Ôºö",np.median(views1))
print("ÊúÄÂ∞èÂÄºÔºö",min(views1))
print("ÊúÄÂ§ßÂÄºÔºö",max(views1))
for i in views:
    if(i!=i):
        values=values+1
print("Áº∫Â§±ÂÄº‰∏™Êï∞:",values)
print("ÂõõÂàÜ‰ΩçÊï∞:",np.percentile(views1, (25, 50, 75), interpolation='midpoint'))
counts = np.bincount(views1)
print("‰ºóÊï∞Ôºö",np.argmax(counts))
```

ÁõíÂõæ

```python
data = pd.DataFrame({"views": views})
data.boxplot()  
pl.ylabel("views")
pl.show()
```

Áõ¥ÊñπÂõæ

```python
def draw_hist(myList,Title,Xlabel,Ylabel,Xmin,Xmax,Ymin,Ymax):
      pl.hist(myList,2300)
      pl.xlabel(Xlabel)
      pl.xlim(Xmin,Xmax)
      pl.ylabel(Ylabel)
      pl.ylim(Ymin,Ymax)
      pl.title(Title)
      pl.show()
data = views
draw_hist(views, 'comment_count', 'comment_count', 'number', min(data), 3000000, 0.0,5000)
```
